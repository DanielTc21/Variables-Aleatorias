La esperanza de $X$ se define como:
$$\mathbb{E}(X) = \sum_{i} x_i \cdot \mathbb{P}(X = x_i)$$

Si $X$ es uniforme con valores de ${(x_1,..,x_n)}$ se tiene que:

$$\mathbb{E}(X) = \frac{x_n + x_1}{2}$$


# Verificamos lo anterior

Definimos entonces que para cada valor de $X$ toma la misma probabilidad dado que la variable es uniforme, por lo que la probabilidad para cada valor es:

$$\mathbb{P}(X=x_i) = \frac{1}{n}$$

Ahora susituimos en la formúla general $\mathbb{E}(X)$ la probabilidad de cada $x_i$, que es $\frac{1}{n}$, tenemos ⁉

$$\mathbb{E}(X) =\sum_{i=1}^{n} {x_i}\cdot\frac{1}{n} $$

Simplificando:

Factorizamos $\frac{1}{n}$ fuera de la sumatoria:

$$\mathbb{E}(X) = \frac{1}{n} \sum_{i=1}^{n} x_i$$

Es decir, que la esperanza es simplemente el promedio de los valores posibles de X


Si $X$ toma valores consecutivos ${x_1,x_2,..,x_n}$ como ocurre en una distribución uniforme, la suma de los valores $\sum_{i=1}^{n} x_i$, se puede escribir como:

$$\sum_{i=1}^{n} x_i = x_1 + x_2 ... + x_n$$

Por lo tanto para una v.u, el valor promedio es simplemente el punto medio entre el valor mínimo $x_1$ y el valor $x_n$

Este resultado se debe a que la variable aleatoria uniforme asigna la misma probabilidad a cada valor, y la esperanza corresponde al promedio de los valores, que en este caso es el promedio de los valores extremos $x_1$ y $x_n$

### 1. Definición de varianza

La **varianza** de una variable aleatoria \( X \) se define como:

$$\text{Var}(X) = \mathbb{E}(X^2) - (\mathbb{E}(X))^2$$


###  Cálculo de  $\mathbb{E}(X^2)$

Para calcular $\mathbb{E}(X^2)$, debemos calcular el valor esperado de los cuadrados de los valores posibles de $X$. Para una variable discreta uniforme, esto es:

$$\mathbb{E}(X^2) = \frac{1}{n} \sum_{i=1}^{n} x_i^2$$

En este caso, los valores de $x_1, x_2, \dots, x_n$ son consecutivos, por lo que podemos usar una fórmula para la suma de los cuadrados de los primeros $n$ enteros consecutivos:

$$\sum_{i=1}^{n} x_i^2 = \frac{n(n+1)(2n+1)}{6}$$

Donde $n = x_n - x_1 + 1$, que es el número total de valores posibles de $X$

### Calculamos la varianza

Con estos resultados, podemos calcular la varianza usando la fórmula general:

$$\text{Var}(X) = \mathbb{E}(X^2) - (\mathbb{E}(X))^2$$

Sustituyendo los valores:

1. $\mathbb{E}(X^2) = \frac{1}{n} \sum_{i=1}^{n} x_i^2$,
2. $\mathbb{E}(X) = \frac{x_1 + x_n}{2}$.

Después de simplificar, obtenemos la fórmula deseada:

$$\text{Var}(X) = \frac{(x_n - x_1 + 1)^2 - 1}{12}$$

### Por lo tanto

Para variables aleatorias discretas uniformes, hemos verificado que la fórmula para la varianza es correcta. Esta fórmula muestra que la varianza depende del rango de los valores posibles de $X$ es decir, $x_n - x_1$ y captura la dispersión de los valores alrededor de su media.
